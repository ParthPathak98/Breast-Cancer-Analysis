```python
import numpy as np 
import pandas as pd 
```

```python
Bdata = pd.read_csv("BreastCancerData.csv")
```

```python
Bdata.head()
```

### Drop unecessary columns.

```python
Bdata.drop(["id"] ,axis = 1 ,  inplace=True)
Bdata.drop(["Unnamed: 32"], axis = 1 , inplace=True)
```

```python
Bdata.shape
```

```python
Bdata.head()
```

### Numerize diagnosis "M" malignant; "B" benign

```python
def mag(str):
    if str =="M":
        return 1   ## setting 1 for Malignant
    else: 
        return 0   ## setting 0 for Benign 
```

```python
Bdata["diagnosis"] = Bdata["diagnosis"].apply(mag)
```

```python
Bdata.tail()
```

## Adding diagnosis to the end of the dataframe


```python
diagnosis = Bdata.diagnosis
```

```python
Bdata.drop(["diagnosis"] , axis = 1  , inplace=True)
```

```python
Bdata["diagnosis"] = diagnosis
```

```python
Bdata.head()
```

## Glimpse of mean feature values wrt Diagnosis

```python
Bdata.groupby("diagnosis").mean()
```

## Graphical representation of data.

```python
x0 = Bdata.diagnosis.value_counts()[0]
x1 = Bdata.diagnosis.value_counts()[1]
print(x0 , x1)
```

```python
import matplotlib.pyplot as plt 
```

```python
labels = "Benign" , "Malingant"
sizes = [x0 , x1]
explode = [0,0.3]
plt.pie(sizes , explode=explode , labels=labels , shadow=True , startangle=90 , autopct='%1.1f%%')
plt.axis("equal")
plt.title("Freq of Diagnosis")
plt.show()
```

## Radius Mean  - Benign vs Malingnant

```python
Bdata1 = Bdata.loc[Bdata["diagnosis"]== 0]
y_Val1 = Bdata1.radius_mean.values.reshape(-1,1)
```

```python
Bdata2 = Bdata.loc[Bdata["diagnosis"] ==1]
y_val2 = Bdata2.radius_mean.values.reshape(-1,1)
```

```python
x1 = np.arange(0,3570,10)
plt.scatter(x1 , y_Val1 , color = "Black" , label="Benign")
plt.axis([0,3570,5,20])
x2 = np.arange(0,2120 , 10)
plt.ylabel("Radius Mean")
plt.scatter(x2 , y_val2 , color = "Red" , label = "Malingnant")
plt.legend()
plt.title("Radius Mean for Benign and Malingnant")
plt.show()
```
